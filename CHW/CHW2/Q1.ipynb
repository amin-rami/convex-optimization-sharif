```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt

P = [0.1, 1, 10, 100]
nums = np.array([0, 4, 2, 2, 3, 0, 4, 5, 6, 6, 4, 1, 4, 4, 0, 1, 3, 4, 2, 0, 3, 2, 0, 1])

constraints = []
opts = []
lamda = cp.Variable(np.size(nums))

fig = plt.figure(figsize=(15, 5))

for i in P:
    objective = cp.Minimize(cp.sum(lamda)
                            - cp.sum(cp.multiply(nums, cp.log(lamda)))
                            + i * cp.sum_squares(lamda[:-1]
                            - lamda[1:]))
                            
    problem = cp.Problem(objective, constraints)
    result = problem.solve()
    opts.append(np.sum(lamda.value) - np.sum(nums * np.log(lamda.value)))
    plt.plot(np.arange(np.size(nums)), lamda.value)

plt.xlabel('t', size = 13)
plt.ylabel(r'$\lambda(t)$', size = 13)
plt.title(r'$\lambda(t)$'+ '  VS  t', size = 13)    
plt.show()    

fig = plt.figure(figsize=(15, 5))
plt.xlabel(r'$\P $', size = 13)
plt.ylabel('Optimal values', size = 13)
plt.title('Optimal values per each '+ r'$\P $', size = 13)
plt.plot(P, opts, 'o', color = 'black')
plt.show()
```
